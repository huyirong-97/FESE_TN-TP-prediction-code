```python
import pandas as pd
import numpy as np
import os
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

from sklearn import tree
model_DecisionTreeRegressor = tree.DecisionTreeRegressor(max_depth=5, random_state=42)
from sklearn import svm
model_SVR = svm.SVR()
from sklearn import neighbors
model_KNeighborsRegressor = neighbors.KNeighborsRegressor(n_neighbors=5)
from sklearn import ensemble
model_RandomForestRegressor = ensemble.RandomForestRegressor(n_estimators=300, max_depth=13, random_state=42)
from sklearn.ensemble import BaggingRegressor
model_BaggingRegressor = BaggingRegressor(n_estimators=400, max_samples=0.66, max_features=0.91, random_state=42)
from xgboost import XGBRegressor
model_XGBRegressor = XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.06, max_depth=2, min_child_weight=7, 
                         subsample=0.95, gamma=0, colsample_bytree=0.45, reg_alpha=0, reg_lambda=0,random_state=42)
```

```python
def get_score(result,y_test,model_name):
    score = r2_score(y_test, result)

    rmse = mean_squared_error(y_test, result) ** 0.5
    MAE = mean_absolute_error(y_test, result)
    MAPE = (abs((y_test - result)/y_test)).mean()*100.0
    
    return rmse,MAE, MAPE,score
```

```python
io = r"C:\Users\Administrator\Desktop\TP_DATA.xlsx"
data_all =pd.read_excel(io)
merge_data = data_all.copy()
train_x = merge_data.iloc[:, :-1]
train_y = merge_data.iloc[:, -1]
```

```python
df_y_test = []
df_y_test_index = []
Feature = []
CART_shuzu = {}
measure = []
CART_pre_result = []
canshu_CART = pd.DataFrame(columns=['RMSE', 'MAE', 'MAPE', 'R2'])
SVM_shuzu = {}
SVM_pre_result = []
canshu_SVR = pd.DataFrame(columns=['RMSE', 'MAE', 'MAPE', 'R2'])
KNN_shuzu = {}
KNN_pre_result = []
canshu_KNN = pd.DataFrame(columns=['RMSE', 'MAE', 'MAPE', 'R2'])
RF_shuzu = {}
RF_pre_result = []
RF_feature = []
canshu_RF = pd.DataFrame(columns=['RMSE', 'MAE', 'MAPE', 'R2'])
BAG_shuzu = {}
BAG_pre_result = []
canshu_Bag = pd.DataFrame(columns=['RMSE', 'MAE', 'MAPE', 'R2'])
XGB_shuzu = {}
XGB_pre_result = []
XGB_feature = []
canshu_xgb = pd.DataFrame(columns=['RMSE', 'MAE', 'MAPE', 'R2'])
```

```python
for i in range(500):
    X_train, X_test, y_train, y_test = train_test_split(train_x, train_y, test_size=0.2,random_state=i)
    df_y_test.append(y_test.values)
    df_y_test_index.append(y_test.index)
    ##########CART#################
    model_DecisionTreeRegressor.fit(X_train,y_train)
    cart_result = model_DecisionTreeRegressor.predict(X_test)
    x,y,z,r = get_score(cart_result,y_test,'CART')
    if r > 0:
        CART_pre_result.append(cart_result)
        CART_shuzu = {'RMSE': x, 'MAE': y, 'MAPE': z, 'R2': r}
        canshu_CART = canshu_CART.append(CART_shuzu, ignore_index=True)
    ##########SVM#################
    model_SVR.fit(X_train,y_train)
    svr_result = model_SVR.predict(X_test)
    x,y,z,r = get_score(svr_result,y_test,'SVR')
    if r > 0:
        SVM_pre_result.append(svr_result)
        SVM_shuzu = {'RMSE': x, 'MAE': y, 'MAPE': z, 'R2': r}
        canshu_SVR = canshu_SVR.append(SVM_shuzu, ignore_index=True)
    ##########KNN#################
    model_KNeighborsRegressor.fit(X_train,y_train)
    KNN_result = model_KNeighborsRegressor.predict(X_test)
    x,y,z,r = get_score(KNN_result,y_test,'KNN')
    if r > 0:
        KNN_pre_result.append(KNN_result)
        KNN_shuzu = {'RMSE': x, 'MAE': y, 'MAPE': z, 'R2': r}
        canshu_KNN = canshu_KNN.append(KNN_shuzu, ignore_index=True)
    ##########RF#################
    model_RandomForestRegressor.fit(X_train,y_train)
    RF_result = model_RandomForestRegressor.predict(X_test)
    x,y,z,r = get_score(RF_result,y_test,'RF')
    if r > 0:
        RF_pre_result.append(RF_result)
        RF_feature.append(model_RandomForestRegressor.feature_importances_)
        RF_shuzu = {'RMSE': x, 'MAE': y, 'MAPE': z, 'R2': r}
        canshu_RF = canshu_RF.append(RF_shuzu, ignore_index=True)
    ##########BAG#################
    model_BaggingRegressor.fit(X_train,y_train)
    Bagging_result = model_BaggingRegressor.predict(X_test)
    x,y,z,r = get_score(Bagging_result,y_test,'Bagging')
    if r > 0:
        BAG_pre_result.append(Bagging_result)
        BAG_shuzu = {'RMSE': x, 'MAE': y, 'MAPE': z, 'R2': r}
        canshu_Bag = canshu_Bag.append(BAG_shuzu, ignore_index=True)
    ##########XGB#################
    model_XGBRegressor.fit(X_train, y_train)
    xgb_result=model_XGBRegressor.predict(X_test)
    x,y,z,r = get_score(xgb_result,y_test,'xgb')
    if r > 0:
        XGB_pre_result.append(xgb_result)
        XGB_feature.append(model_XGBRegressor.feature_importances_)
        XGB_shuzu = {'RMSE': x, 'MAE': y, 'MAPE': z, 'R2': r}
        canshu_xgb = canshu_xgb.append(XGB_shuzu, ignore_index=True)
pre_y_test = pd.DataFrame(df_y_test)
pre_y_test_index = pd.DataFrame(df_y_test_index)
FIN_CART_pre_result = pd.DataFrame(CART_pre_result)
FIN_SVM_pre_result = pd.DataFrame(SVM_pre_result)
FIN_KNN_pre_result = pd.DataFrame(KNN_pre_result)
FIN_RF_pre_result = pd.DataFrame(RF_pre_result)
FIN_BAG_pre_result = pd.DataFrame(BAG_pre_result)
FIN_XGB_pre_result = pd.DataFrame(XGB_pre_result)
RF_feature_wight = pd.DataFrame(RF_feature)
XGB_feature_wight = pd.DataFrame(XGB_feature)
Feature.append(np.mean(RF_feature,axis=0))
Feature.append(np.mean(XGB_feature,axis=0))
Feature_wight = pd.DataFrame(Feature)
Feature_wight.columns = train_x.columns
Feature_wight.index = ['RF', 'XGB']
pre_y_test.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\TP_DATA_y_test.xlsx')
pre_y_test_index.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\TP_DATA_y_test_index.xlsx')
canshu_CART.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\CART_canshu.xlsx')
FIN_CART_pre_result.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\CART_result.xlsx')
canshu_SVR.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\SVM_canshu.xlsx')
FIN_SVM_pre_result.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\SVM_result.xlsx')
canshu_KNN.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\KNN_canshu.xlsx')
FIN_KNN_pre_result.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\KNN_result.xlsx')
canshu_RF.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\RF_canshu.xlsx')
FIN_RF_pre_result.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\RF_result.xlsx')
RF_feature_wight.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\RF_feature_wight.xlsx')
canshu_Bag.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\Bag_canshu.xlsx')
FIN_BAG_pre_result.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\Bag_result.xlsx')
canshu_xgb.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\xgb_canshu.xlsx')
FIN_XGB_pre_result.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\xgb_result.xlsx')
XGB_feature_wight.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\xgb_feature_wight.xlsx')
Feature_wight.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA\new\TP_DATA_FeatureWight.xlsx')
```

```python
def feature_mean_var(input_data):
    feature_mean=pd.DataFrame(np.mean(input_data,axis=0),index=input_data.columns,columns=["mean"])
    feature_var=pd.DataFrame(np.var(input_data,axis=0),index=input_data.columns,columns=["variance"])
    feature_max=pd.DataFrame(np.max(input_data,axis=0),index=input_data.columns,columns=["max"])
    feature_mean_var_max=pd.concat([feature_mean,feature_var,feature_max],axis=1,ignore_index=True)
    return feature_mean_var_max

a = feature_mean_var(canshu_CART)
b = feature_mean_var(canshu_SVR)
c = feature_mean_var(canshu_KNN)
d = feature_mean_var(canshu_RF)
e = feature_mean_var(canshu_Bag)
f = feature_mean_var(canshu_xgb)
Average = pd.concat([a,b,c,d,e,f],axis=1,ignore_index=True)
Average.to_excel(r'C:\Users\Administrator\Desktop\TP_DATA_AllModel_index.xlsx')
```
